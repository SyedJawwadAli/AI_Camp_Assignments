# 6. Write a python program which takes two inputs from user and print them addition

```python
value_1 = int(input("Enter 1st Value :"))

function = input("Enter Function  :")

value_2 = int(input("Enter 2nd Value :"))

if function == "/" :
    print(f"\nDivision : {value_1} {function} {value_2}\nAnswer : " + str( value_1  /  value_2 ))
elif function == "*" :
    print(f"\nMultiplication : {value_1} {function} {value_2}\nAnswer : " + str( value_1  *  value_2 ))
elif function == "+" :
    print(f"\nAddition : {value_1} {function} {value_2}\nAnswer : " + str( value_1  +  value_2 ))
elif function == "-" :
    print(f"\nSubtraction : {value_1} {function} {value_2}\nAnswer : " + str( value_1  -  value_2 ))
else :
    print("\nFor Division : /\nFor Multiplication : *\nFor Addition : +\nFor Subtraction : -")
```

# 7. Write a program which takes 5 inputs from user for different subject's marks, total it and generate mark sheet using grades.

```python
Sub_1 = int(input("Enter Marks in Subject 1 :"))
Sub_2 = int(input("Enter Marks in Subject 2 :"))
Sub_3 = int(input("Enter Marks in Subject 3 :"))
Sub_4 = int(input("Enter Marks in Subject 4 :"))
Sub_5 = int(input("Enter Marks in Subject 5 :"))

print(f"\nSubject 1 : {Sub_1} / 100\nSubject 2 : {Sub_2} / 100\nSubject 3 : {Sub_3} / 100\nSubject 4 : {Sub_4} / 100\nSubject 5 : {Sub_5} / 100\n")

Total = (Sub_1 + Sub_2 + Sub_3 + Sub_4 + Sub_5)
print(f"Total Marks : {Total} / 500")

if Total >= 500 :
    print("\nEnter Correct Marks")
else : 
    Per = Total*100/500
    print(f"\nPercentage : {Per} %\n")
    
    if Per >= 90 and Per <= 100 :
        print("Grade : A One")
    elif Per >= 80 and Per < 90 :
        print("Grade : A")
    elif Per >= 70 and Per < 80 :
        print("Grade : B")
    elif Per >= 60 and Per < 70 :
        print("Grade : C")
    elif Per >= 50 and Per < 60 :
        print("Grade : D")
    elif Per >= 40 and Per < 50 :
        print("Grade : E")
    else :
        print("Grade : F")
```

# 8. Write a program which take input from user and identify that the given number is even or odd?

```python
num = int(input("Enter a Number :"))
mod = num % 2

if mod > 0 :
    print(num, "is an Odd Number.")
else:
    print(num, "is an Even Number.")
```
